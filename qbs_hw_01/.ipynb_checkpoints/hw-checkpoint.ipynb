```python
import pandas as pd

train_data = pd.read_csv('train.csv') 
test_data = pd.read_csv('test.csv') 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

age_for_NaN = train_data.Age.median()
##print(age_for_NaN)

def fill_in_null(column, replace_item):
    for (i, item) in enumerate( column.isnull() ):
        if(item == True):
            column[i] = replace_item

def is_null(column):
    for (i, item) in enumerate( column.isnull() ):
        if(item == True):
            print("NaN")
```

```python
is_null(train_data.Age)
is_null(train_data.fnlwgt)

print(train_data.Age.dtypes)
train_data.fnlwgt = train_data.fnlwgt.astype("float64")
print(train_data.fnlwgt)
```

```python
def normalise(column):
    mean_value = column.mean()
    std_value = column.std()
    for i in range(len(column)):
        column[i] = (column[i]  - mean_value) / std_value

normalise(train_data.fnlwgt)
train_data.fnlwgt = train_data.fnlwgt.round(decimals=2)
print(train_data.fnlwgt)
```

```python
f, axes = plt.subplots(1,6, figsize=(40, 7))
age_pic = sns.distplot(train_data.Age, ax=axes[0])
workclass_pic = sns.countplot(train_data.Workclass, ax=axes[1])

#fnlwgt_vs_target = sns.barplot(x='fnlwgt', y='Target', data=train_data, ax=axes[2])
age_vs_target = sns.barplot(x='Age', y='Target',data=train_data, ax=axes[3])
workclass_vs_target = sns.barplot(x='Workclass', y='Target',data=train_data, ax=axes[4])
```
